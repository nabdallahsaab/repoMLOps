# Dashboard & Application web

Un dashboard est une interface visuelle qui présente les informations clés et les métriques pertinentes d'un projet de data science.

<img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2021/06/39595st.jpeg' width=600>

[Streamlit](https://docs.streamlit.io/) est un framework Python simple et efficace pour créer rapidement des applications web interactives pour la data science.




```python
import pandas as pd
import plotly.express as px


df = pd.read_csv('train.csv', sep=';')
df.head()
```

```python
fig = px.histogram(df.sort_values("education"), x="education", barmode="group")
fig
```

```python
import plotly.graph_objects as go

job = 'management'
data = df[df['job'] == job].groupby('marital').mean().age

# Création de la figure
fig = go.Figure(data=[go.Bar(x=data.index, y=data.values)])

# Mise en forme du graphique
fig.update_layout(title=f'Répartition hommes-femmes en fonction de la profession : {job}',
                  xaxis_title='Genre',
                  yaxis_title='Moyenne d\'âge')
```
